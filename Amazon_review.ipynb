```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Amazon Reviews Data
amazon_review = pd.read_csv("/content/amazon_review.csv")

# 1. Sentiment Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x=amazon_review['overall'], palette='viridis')
plt.title("Review Sentiment Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Categorizing Reviews
amazon_review['Sentiment'] = amazon_review['overall'].apply(lambda x: 'Positive' if x >= 4 else ('Neutral' if x == 3 else 'Negative'))
sentiment_counts = amazon_review['Sentiment'].value_counts()
plt.figure(figsize=(8, 5))
sentiment_counts.plot(kind='bar', color=['green', 'blue', 'red'])
plt.title("Sentiment Categorization")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# 2. Most Reviewed Products
top_products = amazon_review['asin'].value_counts().head(10)
plt.figure(figsize=(10, 5))
top_products.plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Product ASIN")
plt.ylabel("Number of Reviews")
plt.show()

# 3. Review Helpfulness Analysis
amazon_review['Helpfulness_Ratio'] = amazon_review['helpful_yes'] / (amazon_review['total_vote'] + 1)
top_helpful_reviews = amazon_review.nlargest(10, 'Helpfulness_Ratio')
print("Top 10 Most Helpful Reviews:")
print(top_helpful_reviews[['reviewerID', 'asin', 'reviewText', 'Helpfulness_Ratio']])

# 4. Time-Based Trends
amazon_review['reviewTime'] = pd.to_datetime(amazon_review['reviewTime'], format='%d-%m-%Y')
review_trend = amazon_review.groupby(amazon_review['reviewTime'].dt.year)['overall'].count()
plt.figure(figsize=(10, 5))
review_trend.plot(marker='o', color='purple')
plt.title("Review Count Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Amazon Reviews Data
amazon_review = pd.read_csv("/content/amazon_review.csv")

# 1. Sentiment Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x=amazon_review['overall'], palette='viridis')
plt.title("Review Sentiment Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Categorizing Reviews
amazon_review['Sentiment'] = amazon_review['overall'].apply(lambda x: 'Positive' if x >= 4 else ('Neutral' if x == 3 else 'Negative'))
sentiment_counts = amazon_review['Sentiment'].value_counts()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Amazon Reviews Data
amazon_review = pd.read_csv("/content/amazon_review.csv")

# 1. Sentiment Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x=amazon_review['overall'], palette='viridis')
plt.title("Review Sentiment Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Categorizing Reviews
amazon_review['Sentiment'] = amazon_review['overall'].apply(lambda x: 'Positive' if x >= 4 else ('Neutral' if x == 3 else 'Negative'))
sentiment_counts = amazon_review['Sentiment'].value_counts()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Amazon Reviews Data
amazon_review = pd.read_csv("/content/amazon_review.csv")

# 1. Sentiment Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x=amazon_review['overall'], palette='viridis')
plt.title("Review Sentiment Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Categorizing Reviews
amazon_review['Sentiment'] = amazon_review['overall'].apply(lambda x: 'Positive' if x >= 4 else ('Neutral' if x == 3 else 'Negative'))
sentiment_counts = amazon_review['Sentiment'].value_counts()
plt.figure(figsize=(8, 5))
sentiment_counts.plot(kind='bar', color=['green', 'blue', 'red'])
plt.title("Sentiment Categorization")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# 2. Most Reviewed Products
top_products = amazon_review['asin'].value_counts().head(10)
plt.figure(figsize=(10, 5))
top_products.plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Product ASIN")
plt.ylabel("Number of Reviews")
plt.show()

# 3. Review Helpfulness Analysis
amazon_review['Helpfulness_Ratio'] = amazon_review['helpful_yes'] / (amazon_review['total_vote'] + 1)
top_helpful_reviews = amazon_review.nlargest(10, 'Helpfulness_Ratio')
print("Top 10 Most Helpful Reviews:")
print(top_helpful_reviews[['reviewerID', 'asin', 'reviewText', 'Helpfulness_Ratio']])

# 4. Time-Based Trends
amazon_review['reviewTime'] = pd.to_datetime(amazon_review['reviewTime'], format='%d-%m-%Y')
review_trend = amazon_review.groupby(amazon_review['reviewTime'].dt.year)['overall'].count()
plt.figure(figsize=(10, 5))
review_trend.plot(marker='o', color='purple')
plt.title("Review Count Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Reviews")
plt.show()

# 5. Product Performance Analysis
product_ratings = amazon_review.groupby('asin')['overall'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
product_ratings.head(10).plot(kind='bar', color='gold')
plt.title("Top 10 Highest Rated Products")
plt.xlabel("Product ASIN")
plt.ylabel("Average Rating")
plt.show()

# 6. Customer Loyalty & Retention
reviewer_counts = amazon_review['reviewerID'].value_counts()
repeat_reviewers = reviewer_counts[reviewer_counts > 1].count()
print(f"Number of Repeat Reviewers: {repeat_reviewers}")

# 7. Competitor Benchmarking
brand_performance = amazon_review.groupby('asin')['overall'].mean()
print("Top 10 Brands by Average Rating:")
print(brand_performance.nlargest(10))

# 8. Impact of Review Helpfulness
amazon_review['Review_Length'] = amazon_review['reviewText'].astype(str).apply(len)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=amazon_review['Review_Length'], y=amazon_review['Helpfulness_Ratio'], alpha=0.5)
plt.title("Review Length vs. Helpfulness")
plt.xlabel("Review Length")
plt.ylabel("Helpfulness Ratio")
plt.show()

# 9. Seasonal Trends in Reviews
amazon_review['Review_Month'] = amazon_review['reviewTime'].dt.month
monthly_reviews = amazon_review.groupby('Review_Month')['overall'].count()
plt.figure(figsize=(10, 5))
monthly_reviews.plot(kind='bar', color='cyan')
plt.title("Monthly Review Trends")
plt.xlabel("Month")
plt.ylabel("Number of Reviews")
plt.show()
```
